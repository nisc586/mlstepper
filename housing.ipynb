```python
import numpy as np
from linreg import LinearRegression
import matplotlib.pyplot as plt
```

```python
housing = np.loadtxt("datasets/housing.data")
housing.shape
```

```python
types = ["CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"]

def get_feature(s):
    idx = types.index(s)
    return housing[:, idx]
```

```python
rooms = get_feature("RM")
price = get_feature("MEDV")
```

```python
model = LinearRegression(rooms, price, regressors=[lambda x: x])
model.solve()
model.show()
```
